```python
#All necessary Imports
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Dataset 
df = pd.read_csv("house_data.csv")

#Extracting necessary feature
df = df[["price","sqft_living"]]
display(df)
```

```python
#Feature selection and assigning it to variables
X = df[["sqft_living"]].values #Independent Variable
Y = df[["price"]].values #Dependent Variable (as price depends on the size of property)

df.plot(kind="scatter", x="sqft_living", y="price")
```

```python
#Training a linear regression model 
model = LinearRegression()
model.fit(X,Y)
```

```python
#Testing the model on unseen data
#For given size predict the price based on training data
X_new = [[1905]]
print(model.predict(X_new))
```
